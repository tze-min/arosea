```python
doc = nlp("Bakalia, Kotwali, Chandgaon, Pahartali, Hathazari, Halishahar, Patiya, Satkania areas (Chittagong district, Chittagong province), Cox's Bazar district (Chittagong province)")
for ent in doc.ents:
    if ent.label_ == "GPE":
        print(ent.text, ent.label_)
```

```python
import spacy
nlp = spacy.load("en_core_web_lg")

def getGPE(string):
    locations = []
    doc = nlp(string)
    for ent in doc.ents:
        if ent.label_ == "GPE":
            locations.append(ent.label_)
    return locations
```

```python
eg2 = eg.apply(lambda x: pd.Series(getGPE(x), index=[""]), result_type="expand")
eg2
```

```python
def parseToColumn(row):
    disno = row["Dis No"]
    lst = geo.geoparse(row["Location"]) # ls = [{kv, kdict[lat], kdict[lon]},{},{}]
    res = []
    
    for d in lst:
        geodata = d.get("geo")
        if geodata:
            lon, lat = d["geo"]["lon"], d["geo"]["lat"]
            location = d["word"]
            country = d["country_predicted"]
            res.append([disno, location, country, lon, lat])
        else:
            continue
    return res
```
